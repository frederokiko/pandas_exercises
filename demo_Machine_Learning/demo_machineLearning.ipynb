# ML AI

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel('./Salary.xlsx')
```

```python
df.head()
```

```python
X = df[["annee Experience"]]
```

```python
y = df.Salaire
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.score(X, y)
```

```python
annexp = 10

f"Pour {annexp} annee d'experience le salaire est de $ {model.predict([[annexp]])[0]}"
```

```python
import numpy as np
annee_exp = np.arange(0,9)

```

```python
annee_exp.reshape(-1,1)
```

```python
prediction=model.predict(annee_exp.reshape(-1,1))
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
plt.plot(annee_exp,prediction,color='red')
plt.show()
```
